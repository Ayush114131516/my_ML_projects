## Recommender System

```python
numdf=df.groupby('Book-Title').count()['Book-Rating'].reset_index()
numdf.rename(columns={'Book-Rating':'rating_count'},inplace=True)
numdf.head(3)
```

```python
avgdf=df.groupby('Book-Title').mean()['Book-Rating'].reset_index()
avgdf.rename(columns={'Book-Rating':'rating_avg'},inplace=True)
avgdf.head(3)
```

```python
popularitydf=numdf.merge(avgdf,on='Book-Title')
popularitydf.head(3)
```

```python
popularitydf.shape
```

```python
pop30df=popularitydf[popularitydf['rating_count']>=200].sort_values('rating_avg',ascending=False).head(30)
pop30df
```

```python
pop30df=pop30df.merge(books,on='Book-Title').drop_duplicates('Book-Title')[['Book-Title','Book-Author','Year-Of-Publication','Image-URL-L','rating_count','rating_avg']]
pop30df.head(5)
```

```python
df2=df.groupby('User-ID').count()['Book-Rating']

x=df2[df2>=200].index #boolean indexing

filtered=df[df['User-ID'].isin(x)]
```

```python
y=filtered.groupby('Book-Title').count()['Book-Rating']>=50
famous_books=y[y].index
famous_books
```

```python
final=filtered[filtered['Book-Title'].isin(famous_books)].drop_duplicates()
```

```python
table=final.pivot_table(index='Book-Title',columns='User-ID',values='Book-Rating').fillna(0)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
score_of_similarity=cosine_similarity(table)
```

Steps:

```python
np.where(table.index=='2nd Chance')[0][0]
```

```python
sorted(list(enumerate(score_of_similarity[0])),key=lambda x:x[1],reverse=True)[1:5]
```

THE Function

```python
def recommend(book_name):
  index=np.where(table.index==book_name)[0][0]
  similar_books=sorted(list(enumerate(score_of_similarity[index])),key=lambda x:x[1],reverse=True)[1:5]

  for i in similar_books:
    print(table.index[i[0]])
```

```python
recommend('Message in a Bottle')
```

```python
recommend('1984')
```
