```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv('/content/Churn_Modelling.csv')
```

```python
X=df.iloc[:,3:-1] #extracting the independent columns
y=df.iloc[:,-1].values.ravel() #ertracting the target column
```

```python
y
```

```python
# the categorical data
print(X['Geography'].value_counts())
print(X['Gender'].value_counts())
print(X['HasCrCard'].value_counts())
print(X['IsActiveMember'].value_counts())
```

```python
X.describe()
```

```python
X.isnull().sum()
```

```python
X=pd.get_dummies(X,drop_first=True)
X.head()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,stratify=y,random_state=0)
```

```python
# preprrocessing
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X_train=sc.fit_transform(X_train)
X_test=sc.transform(X_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier()
rf.fit(X_train ,  y_train)
```

```python
clac_y=rf.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix
print(confusion_matrix(y_test,clac_y))
print(accuracy_score(y_test,clac_y))
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(X_train,y_train)
clac_y2=lr.predict(X_test)
```

```python
from sklearn.metrics import r2_score
print(r2_score(y_test,clac_y2))
```
